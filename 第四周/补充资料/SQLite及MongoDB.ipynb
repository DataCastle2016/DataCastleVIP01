### SQLite

特点：

不需要一个单独的服务器进程或操作的系统（无服务器的）。

SQLite 不需要配置，这意味着不需要安装或管理。

一个完整的 SQLite 数据库是存储在一个单一的跨平台的磁盘文件。

SQLite 是非常小的，是轻量级的，完全配置时小于 400KiB，省略可选功能配置时小于250KiB。

SQLite 是自给自足的，这意味着不需要任何外部的依赖。

SQLite 事务是完全兼容 ACID 的，允许从多个进程或线程安全访问。

SQLite 支持 SQL92（SQL2）标准的大多数查询语言的功能。

SQLite 使用 ANSI-C 编写的，并提供了简单和易于使用的 API。

SQLite 可在 UNIX（Linux, Mac OS-X, Android, iOS）和 Windows（Win32, WinCE, WinRT）中运行。
文件型数据库

常用于应用程序存储数据，比如浏览器存储用户的资料等

GUI： DB Browser for SQLite；访问SQLite就是访问一个文件，使用这个前端的程序来访问

下载地址:https://sqlitebrowser.org/dl/

Python中可以使用sqlite3来访问

SQLite的操作使用的是SQL语句进行操作。

#### 数据类型
NULL	：值是一个 NULL 值。

INTEGER	：值是一个带符号的整数，根据值的大小存储在 1、2、3、4、6 或 8 字节中。

REAL	：值是一个浮点值，存储为 8 字节的 IEEE 浮点数字。

TEXT	：值是一个文本字符串，使用数据库编码（UTF-8、UTF-16BE 或 UTF-16LE）存储。

BLOB	：值是一个 blob 数据，完全根据它的输入存储。

### Mongodb
MongoDB 是由C++语言编写的，是一个基于分布式文件存储的开源数据库系统。

在高负载的情况下，添加更多的节点，可以保证服务器性能。

MongoDB 旨在为WEB应用提供可扩展的高性能数据存储解决方案。

MongoDB 将数据存储为一个文档，数据结构由键值(key=>value)对组成。MongoDB 文档类似于 JSON 对象。字段值可以包含其他文档，数组及文档数组。

特点：

NoSQL数据库，不使用SQL语句进行检索

MongoDB存储文件的类型类似于JSON格式

GUI：Robo 3T

下载地址：https://robomongo.org/

Python：PyMongo，来进行数据库的读取（需要安装）

### PyMongo
安装：pip install pymongo



```python
#导入pymongo
import pymongo

#连接MongoDB
myclient = pymongo.MongoClient("mongodb://localhost:27017/")

#创建数据库
mydb = myclient["demo"]

#创建集合
mycol = mydb["sites"]

#数据
mydict = { "name": "张三", "sex": "男", "age": "12" }

#插入一条数据
mycol.insert_one(mydict)

#插入多条数据
mydict = [
    { "name": "张三", "sex": "男", "age": "12" },
    { "name": "李四", "sex": "男", "age": "15" },
    { "name": "王五", "sex": "女", "age": "13" },
    { "name": "赵六", "sex": "女", "age": "17" }
]
mycol.insert_many(mydict)
```

```python
#查询数据

#查询一条数据
x = mycol.find_one()
print(x)

#查询多条数据
for x in mycol.find():
    print(x)
```

```python
#修改数据
myquery = { "name": "张三" }
newvalues = { "$set": { "name": "张二" } }

#update_one() 方法只能修匹配到的第一条记录，如果要修改所有匹配到的记录，可以使用 update_many()。
mycol.update_one(myquery, newvalues)
 
# 输出修改后的  "sites"  集合
for x in mycol.find():
    print(x)
```

```python
#数据排序
mydoc = mycol.find().sort("age")
for x in mydoc:
    print(x)
```

```python
#删除数据
myquery = { "name": "张三" }
 
mycol.delete_one(myquery)
 
# 删除后输出
for x in mycol.find():
    print(x)
```
